```python
poverty_df = poverty_df[['name','B17001001','B17001002','B17001031']]
```

```python
poverty_df = poverty_df.rename(columns={"B17001001": "Total Poverty Population" , 
                                      "B17001002" :"Income below poverty level",
                                      "B17001031": "Income above poverty level", 
                                    })
poverty_df.head()
```

```python
income_df = income_df[['name','B19001001','B19001002','B19001003','B19001004','B19001005','B19001006','B19001007','B19001008','B19001009','B19001010','B19001011','B19001012','B19001013','B19001014','B19001015','B19001016']]
```

```python
income_df = income_df.rename(columns={"B19001001": "Total Income Population" ,
                                      "B19001002": "Less than $10,000",
                                      "B19001003": "$10,000 to $14,999",
                                      "B19001004": "$15,000 to $19,999",
                                      "B19001005": "$20,000 to $24,999",
                                      "B19001006": "$25,000 to $29,999",
                                      "B19001007": "$30,000 to $34,999",
                                      "B19001008": "$35,000 to $39,999",
                                      "B19001009": "$40,000 to $44,999",
                                      "B19001010": "$45,000 to $49,999",
                                      "B19001011": "$50,000 to $59,999",
                                      "B19001012": "$60,000 to $74,999",
                                      "B19001013": "$75,000 to $99,999",
                                      "B19001014": "$100,000 to $124,999",
                                      "B19001015": "$125,000 to $149,999",
                                      "B19001016": "$150,000 to $199,999",
                                      "B19001017": "$200,000 or More"
                                        })
income_df.head()
```

```python
both_df = pd.merge(poverty_df, income_df, on='name')
```

```python
both_df.head()
```

```python
#Connect to local database
rds_connection_string = f"root:{password}@localhost/census_db"
engine = create_engine(f'mysql://{rds_connection_string}')
```

```python
#Check tables
engine.table_names()
```

```python
#Load df into table in mysql
both_df.to_sql(name='census', con=engine, if_exists='append', index=False)
```

```python
#Preview table to verify load worked
pd.read_sql_query('select * from census', con=engine).head()
```
